```python
import requests 
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.britannica.com/topic/list-of-countries-1993160')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
containers = soup.find_all("ul", {"class":"with-dots"})
links = [link.findChildren("li") for link in containers]
```

```python
result = []
for countries in links:
    for country in countries:
        result.append(country)
result = [country.get_text() for country in result]
```

```python
print(len(result))
```

```python
import pandas as pd
df = pd.DataFrame({'name':result})
df.to_csv("countries_webscrape")
```
